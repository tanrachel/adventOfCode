```python
import cmath  # Complex math module for handling complex roots
import math 
def solve_quadratic(a, b, c):
    # Calculate the discriminant
    delta = b**2 - 4*a*c

    # Check if the discriminant is non-negative
    # print(delta)
    if delta >= 0:
        # Real roots
        root1 = (-b + delta**0.5) / (2*a)
        root2 = (-b - delta**0.5) / (2*a)
        return float(root1), float(root2)
    else:
        # No real roots
        return None

def getDiffOfRoot(root1,root2):
    root1 = float(root1)
    root2 = float(root2)
    if root1 > root2:
        minRoot = root2 
        maxRoot = root1 
    else: 
        minRoot = root1 
        maxRoot = root2 
    upperBound = math.ceil(maxRoot-1)
    lowerBound = math.floor(minRoot+1)
    return upperBound - lowerBound + 1

time = [7,15,30]
dist = [9,40,200]
time = [40,92,97,90]
dist = [215,1064,1505,1100]
total = 1
for i in range(0,len(time)):
    root1, root2 = solve_quadratic(1,-time[i],dist[i])
    getDiff = getDiffOfRoot(root1,root2)
    # print(getDiff, root1,root2)
    total *= getDiff
print(total)
# root1,root2 = solve_quadratic(1,-7,9)
# print(root1,root2)
# print(getDiffOfRoot(root1,root2))

# root1,root2 = solve_quadratic(1,-15,40)
# print(root1,root2)

# print(getDiffOfRoot(root1,root2))

# part 2

root1, root2 = solve_quadratic(1,-40929790,215106415051100)
getDiff = getDiffOfRoot(root1,root2)
print(getDiff)
```
