```python
f = open("day8.txt", "r")
day8 = (f.read()).split("\n")

instructions = day8[0]
day8 = day8[2:]
nodeMap = {}
starterNodes = []
for i in range(0,len(day8)):
    node = day8[i].split("=")
    nodeBranch = node[1].replace("(","").replace(")","").replace(" ","").strip(" ").split(",")
    nodeHead = node[0].strip(" ")
    if nodeHead[-1] == "A":
        starterNodes.append(nodeHead)
    nodeMap[nodeHead] = nodeBranch

instructions = [x for x in instructions]

nextNodes = []
nodeLRs = []
for starterNode in starterNodes:
    nextNodes.append(starterNode)
    nodeLRs.append(nodeMap[starterNode])
stepCounter = 0
stepTracker = []
nodeReachedTracker = {}
def isAllInDest(array, stepCounter):
    for i in array:
        if i[-1] != "Z":
            return False 
        else: 
            if i not in nodeReachedTracker:
                nodeReachedTracker[i] = []
            reachedArray = nodeReachedTracker[i]
            reachedArray.append(stepCounter)
            nodeReachedTracker[i] = reachedArray
    return True
print("starter nodes: ", nextNodes)
while len(instructions) != 0:
    step = instructions.pop(0)
    stepTracker.append(step)
    
    tempNodesTracker = []
    tempNodeLRs = []
    while len(nextNodes) != 0:
        nextNode = nextNodes.pop(0)
        nodeLR = nodeLRs.pop(0)
        if step == "R":
            nextNode = nodeLR[1]
        else: 
            nextNode = nodeLR[0]

        tempNodesTracker.append(nextNode) 
        tempNodeLRs.append(nodeMap[nextNode])
    stepCounter += 1 
    nextNodes = tempNodesTracker
    nodeLRs = tempNodeLRs
    checkIfWeReached = isAllInDest(nextNodes, stepCounter)
    # print(nextNodes, step, stepCounter, checkIfWeReached)
    if (not checkIfWeReached) and len(instructions)==0:
        while len(stepTracker) != 0:
            instructions.append(stepTracker.pop(0))
    # print(len(nodeReachedTracker), len(starterNodes))
    if len(nodeReachedTracker) == len(starterNodes):
        print("I reached ZZZ: ", nextNode)
        break
print(nodeReachedTracker) 
print(stepCounter)
```
