```python
f = open("day1_1.txt", "r")
data = (f.read()).split("\n")
sumOfTotal = 0 
for row in data:
    firstIndex = 0
    for i in range(0,len(row)):
        if (row[i]).isnumeric():
            firstIndex = i
            break 
    secondIndex = 0
    for j in range(len(row)-1, i-1,-1):
        if (row[j]).isnumeric():
            secondIndex = j
            break 
    sumOfTotal += int(row[firstIndex]+row[secondIndex])
print(sumOfTotal)
```

```python
validNumbers = {"one":"1",
                "two":"2",
                "three":"3",
                "four":"4",
                "five":"5",
                "six":"6",
                "seven":"7",
                "eight":"8",
                "nine":"9"}
lettersToTrack = {"o":["one"],
                  "t":["two","three"],
                  "f":["four","five"],
                  "s":["six","seven"],
                  "e":["eight"],
                  "n":["nine"]}
totalSum = 0
for row in data:
    validValues = list()
    for i in range(0,len(row)):
        if row[i] in lettersToTrack:
            valuesToCheck = lettersToTrack[row[i]]
            for value in valuesToCheck:
                if row[i:i+len(value)] == value:
                    validValues.append(validNumbers[row[i:i+len(value)]]) 
        if row[i].isnumeric():
            validValues.append(row[i])
    finalValue = validValues[0] + validValues[-1]
    totalSum += int(finalValue)
print(totalSum)    
```

```python
f = open("day3_1.txt", "r")
day3_data = (f.read()).split("\n")

def getSurroundingValues(row,col,data):
    surroundingLoc = list()
    for r in [-1,0,1]:            
        for c in [-1,0,1]:
            if row == 0 and r == -1:
                continue 
            if row == len(data)-1 and r == 1:
                continue 
            if col == 0 and c == -1:
                continue 
            if col == len(data[row])-1 and c == 1:
                continue
            
            loc = [row + r,col + c]
            # print("loc: ",loc, " row: ",row," col: ",col)
            surroundingLoc.append(loc)
    return surroundingLoc
            

def checkForSymbols(row,col,data,gearTracker):
    loc = getSurroundingValues(row,col,data)
    for l in loc:
        r = l[0]
        c = l[1]
        value = data[r][c]
        if not (value.isdigit()):
            if value == "*":
                key = str(r)+","+str(c)
                # print("key",key," - ", str(r)," - ",str(c))
                if key not in gearTracker:
                    # print("I should be here: ",key)
                    gearTracker[key] = list()
                return True,key
    return False,""

totalSum = 0 
gearTracker = {}     
for rowNum in range(0, len(day3_data)):
    
    valuesToCheck = list(day3_data[rowNum])
    # print("valuesToCheck: ",valuesToCheck)
    valueBuilder = ""
    isValid = False
    isNumber = False
    gearKey = ""
    for columnNum in range(0, len(day3_data[rowNum])):
        currentVal = day3_data[rowNum][columnNum]
        if currentVal.isnumeric():
            isNumber = True 
        else:
            isNumber = False
        if isNumber:
            valueBuilder += currentVal  
            hasSymbol, gK = checkForSymbols(rowNum,columnNum,day3_data,gearTracker)
            if hasSymbol:
                isValid = True 
                gearKey = gK
        else:
            if isValid:
                gearTracker[gearKey].append(int(valueBuilder))
                isValid = False 
            gearKey = ""
            valueBuilder = ""
    if isValid:
        gearTracker[gearKey].append(int(valueBuilder))
for key in gearTracker:
    gears = gearTracker[key]
    if len(gears)==2:
        totalSum += gears[0]*gears[1]
print(totalSum)
```
